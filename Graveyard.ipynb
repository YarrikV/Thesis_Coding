# Loading dataset from raw config files
Very slow approach, faster is dumping this dataset and loading it with pickle when you need it.

Origin: `NN.ipynb`

```python
# Load general dataset
import pickle

root = os.path.join("thesis", "data", "configs")


for L in [16,24,32,40,48,56,64]:
  saveDataSet(L)

def saveDataSet(L):
  train_data = load_dataSet(root, L, Tc0, max_data_Size, train_on_vortex)f

  filename = f"dataset_L={L}"
  saveDataSet_for_L(filename, train_data, L)


def saveDataSet_for_L(filename, dataSet, L):
  %cd ./thesis/
  with open(filename, "wb") as dump_file:
    pickle.dump(dataSet, dump_file)
  %cd ./../
  print(f"loaded dataset L={L} saved")
```

# Training data Analysis
Only interesting to do once on the training data.

Origin: `NN.ipynb`

```python
L_range = [_ for _ in range(8,64+1,8)]
for L_ in L_range:
    classes_labels = np.array([0,1])
    classes = train_data.labels.int().squeeze().bincount().cpu()

    print(classes)
    print(f"Class distribution: {100*(classes[0]/classes.sum()):.0f}% / {100*(classes[1]/classes.sum()):.0f}%")
    width=1
    fig, ax = plt.subplots()

    ax.bar(classes_labels, classes/classes.sum()*100, width-.1, label="class distribution")
    ax.set_xticks(classes_labels)
    ax.set_ylabel(f"Percentage (of {classes.sum()} samples)")
    ax.set_xlabel("Classes")
    ax.set_title(f"Class distribution of configurations of size {L_}")
    plt.show()
```
